```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from typing import List, Dict, Tuple
from dataclasses import dataclass
import warnings

# Configura√ß√µes globais
warnings.filterwarnings("ignore")
sns.set(style="whitegrid", palette="pastel")
plt.rcParams["figure.dpi"] = 120
plt.rcParams['font.size'] = 10
PALETA_CORES = sns.color_palette("Set2")

# Constantes
URL_DADOS: str = "https://archive.ics.uci.edu/ml/machine-learning-databases/statlog/german/german.data"
NOMES_COLUNAS: List[str] = [
    "status_conta", "duracao_meses", "historico_credito", "proposito", "valor_credito", "saldo_poupanca",
    "tempo_emprego", "taxa_parcela", "estado_civil_sexo", "outros_devedores", "tempo_residencia",
    "propriedade", "idade", "outros_planos", "moradia", "creditos_existentes",
    "emprego", "dependentes", "telefone", "trabalhador_estrangeiro", "risco_credito"
]

@dataclass
class ConfiguracaoEDA:
    variaveis_numericas: List[str]
    variaveis_categoricas: List[str]
    variavel_alvo: str = "risco_credito"


class AnalisadorDadosCredito:
    """
    Classe para an√°lise explorat√≥ria de dados de risco de cr√©dito alem√£o.
    
    Realiza an√°lise completa incluindo:
    - Vis√£o geral dos dados
    - Distribui√ß√£o da vari√°vel alvo
    - An√°lise de vari√°veis num√©ricas e categ√≥ricas
    - Correla√ß√µes entre vari√°veis
    - Detec√ß√£o de outliers
    """
    
    def __init__(self, url: str = URL_DADOS) -> None:
        self.df: pd.DataFrame = self._carregar_dados(url)
        self.config: ConfiguracaoEDA = self._inicializar_configuracao()
    
    def _carregar_dados(self, url: str) -> pd.DataFrame:
        """Carrega e prepara os dados, convertendo a vari√°vel alvo para bin√°ria."""
        df = pd.read_csv(url, sep=' ', header=None, names=NOMES_COLUNAS)
        df["risco_credito"] = df["risco_credito"].map({1: 1, 2: 0})  # 1=ruim, 0=bom
        return df

    def _inicializar_configuracao(self) -> ConfiguracaoEDA:
        """Identifica automaticamente os tipos de vari√°veis para an√°lise."""
        variaveis_numericas = self.df.select_dtypes(include=np.number).drop(columns="risco_credito").columns.tolist()
        variaveis_categoricas = self.df.select_dtypes(exclude=np.number).columns.tolist()
        return ConfiguracaoEDA(variaveis_numericas, variaveis_categoricas)

    def executar_analise_completa(self) -> None:
        """Orquestra todo o fluxo de an√°lise com explica√ß√µes detalhadas."""
        print("üìä IN√çCIO DA AN√ÅLISE EXPLORAT√ìRIA DE DADOS DE CR√âDITO üìä\n")
        
        self._mostrar_visao_geral()
        self._plotar_distribuicao_alvo()
        self._analisar_variaveis_numericas()
        self._analisar_variaveis_categoricas()
        self._analisar_correlacoes()
        self._detectar_outliers()
        
        print("\n‚úÖ AN√ÅLISE CONCLU√çDA COM SUCESSO!")

    def _mostrar_visao_geral(self) -> None:
        """Exibe informa√ß√µes fundamentais sobre o conjunto de dados."""
        print("üîç VIS√ÉO GERAL DO CONJUNTO DE DADOS")
        print("="*50)
        print(f"üìå Total de clientes analisados: {self.df.shape[0]:,}")
        print(f"üìå Caracter√≠sticas dispon√≠veis: {self.df.shape[1]}")
        
        print("\nüìä TIPOS DE DADOS PRESENTES:")
        print(self.df.dtypes.value_counts().to_string())
        
        print("\n‚ùó VALORES AUSENTES (FALTANTES):")
        missing = self.df.isnull().sum()[lambda x: x > 0]
        print(missing.to_string() if not missing.empty else "Nenhum valor ausente encontrado!")
        
        print("\nüìà ESTAT√çSTICAS DESCRITIVAS (VARI√ÅVEIS NUM√âRICAS):")
        print(self.df[self.config.variaveis_numericas].describe().T.to_string())
        
        print("\n" + "="*50 + "\n")

    def _plotar_distribuicao_alvo(self) -> None:
        """Cria visualiza√ß√£o da distribui√ß√£o da vari√°vel alvo com explica√ß√£o detalhada."""
        plt.figure(figsize=(10, 6))
        
        # Gr√°fico de barras
        ax = sns.countplot(data=self.df, x=self.config.variavel_alvo, palette=PALETA_CORES[:2])
        
        # C√°lculos para anota√ß√µes
        total = len(self.df)
        contagens = self.df[self.config.variavel_alvo].value_counts()
        porcentagens = (contagens / total * 100).round(1)
        
        # Adicionando anota√ß√µes
        for i, p in enumerate(ax.patches):
            ax.annotate(f"{contagens[i]}\n({porcentagens[i]}%)", 
                       (p.get_x() + p.get_width() / 2., p.get_height()),
                       ha='center', va='center', xytext=(0, 10), 
                       textcoords='offset points', fontsize=10)

        plt.title("Distribui√ß√£o do Risco de Cr√©dito\n(0 = Bom Pagador, 1 = Mau Pagador)", pad=20, fontsize=12)
        plt.xticks([0, 1], ["Bons Pagadores", "Maus Pagadores"], fontsize=10)
        plt.ylabel("N√∫mero de Clientes", fontsize=10)
        plt.xlabel("")
        plt.tight_layout()
        plt.show()
        
        print("""
üìå **EXPLICA√á√ÉO DO GR√ÅFICO DE DISTRIBUI√á√ÉO DO RISCO**:

Este gr√°fico de barras mostra como os clientes est√£o distribu√≠dos entre bons e maus pagadores:

1. **Eixo X**: 
   - 0 representa clientes com cr√©dito BOM (pagam em dia)
   - 1 representa clientes com cr√©dito RUIM (inadimplentes)

2. **Eixo Y**: 
   - Mostra a quantidade de clientes em cada categoria

3. **Porcentagens**:
   - Indicam a propor√ß√£o de cada grupo em rela√ß√£o ao total
   - Exemplo: Se mostra 70% bons e 30% ruins, significa que para cada 10 clientes, 
     7 pagam em dia e 3 t√™m hist√≥rico de inadimpl√™ncia

4. **Import√¢ncia**:
   - Ajuda a entender se o conjunto de dados est√° balanceado
   - Em cr√©dito, √© normal ter mais bons que maus pagadores
   - Desbalanceamentos extremos podem afetar modelos de machine learning

üí° **O que observar**:
- Qual a propor√ß√£o entre bons e maus pagadores?
- H√° desbalanceamento extremo (ex: menos de 10% em uma categoria)?
- A propor√ß√£o reflete a realidade do mercado de cr√©dito?
""")

    def _analisar_variaveis_numericas(self) -> None:
        """Analisa cada vari√°vel num√©rica com histograma e boxplot."""
        print("\nüìä AN√ÅLISE DAS VARI√ÅVEIS NUM√âRICAS")
        print("="*50)
        
        for var in self.config.variaveis_numericas:
            plt.figure(figsize=(14, 6))
            plt.suptitle(f"An√°lise da Vari√°vel: {var}", y=1.02, fontsize=12)
            
            # Histograma
            plt.subplot(1, 2, 1)
            sns.histplot(
                data=self.df, x=var, hue=self.config.variavel_alvo,
                kde=True, element="step", palette=PALETA_CORES[:2],
                bins=30, alpha=0.6
            )
            plt.title(f"Distribui√ß√£o de {var} por Risco", pad=10, fontsize=11)
            plt.xlabel(var, fontsize=10)
            plt.ylabel("Densidade", fontsize=10)
            plt.legend(title="Risco", labels=["Bom", "Ruim"])
            
            # Boxplot
            plt.subplot(1, 2, 2)
            sns.boxplot(
                data=self.df, x=self.config.variavel_alvo, y=var,
                palette=PALETA_CORES[:2], showfliers=False,
                width=0.5
            )
            plt.title(f"Compara√ß√£o por Risco", pad=10, fontsize=11)
            plt.xlabel("Risco de Cr√©dito", fontsize=10)
            plt.ylabel(var, fontsize=10)
            plt.xticks([0, 1], ["Bom", "Ruim"])
            
            plt.tight_layout()
            plt.show()
            
            print(f"""
üìå **AN√ÅLISE DA VARI√ÅVEL NUM√âRICA: {var.upper()}**

1. **GR√ÅFICO DA ESQUERDA (Histograma)**:
   - Mostra como os valores de {var} est√£o distribu√≠dos
   - As cores diferenciam bons (azul) e maus (laranja) pagadores
   - A linha suave (KDE) mostra a densidade da distribui√ß√£o
   - Ideal para ver padr√µes como:
     * Valores mais comuns (picos do gr√°fico)
     * Assimetria (dados concentrados √† esquerda ou direita)
     * Diferen√ßas entre bons e maus pagadores

2. **GR√ÅFICO DA DIREITA (Boxplot)**:
   - Resume a distribui√ß√£o atrav√©s de quartis
   - A caixa mostra onde est√£o 50% dos valores (Q1 a Q3)
   - A linha na caixa √© a MEDIANA (valor do meio)
   - Os "bigodes" mostram o intervalo normal dos dados
   - √ötil para comparar:
     * Diferen√ßa nas medianas entre grupos
     * Dispers√£o dos valores (tamanho da caixa)
     * Posi√ß√£o relativa dos grupos

üí° **O QUE OBSERVAR PARA {var}**:
- Maus pagadores tendem a ter valores maiores ou menores?
- H√° sobreposi√ß√£o significativa entre os grupos?
- Existem diferen√ßas claras nas distribui√ß√µes?
- H√° valores extremos que podem distorcer a an√°lise?
""")

    def _analisar_variaveis_categoricas(self) -> None:
        """Analisa cada vari√°vel categ√≥rica com gr√°fico de barras."""
        print("\nüìä AN√ÅLISE DAS VARI√ÅVEIS CATEG√ìRICAS")
        print("="*50)
        
        for var in self.config.variaveis_categoricas:
            plt.figure(figsize=(12, 6))
            
            # Ordena por frequ√™ncia e limita a 15 categorias para clareza
            contagens = self.df[var].value_counts().nlargest(15)
            df_filtrado = self.df[self.df[var].isin(contagens.index)]
            
            # Gr√°fico de barras
            ax = sns.countplot(
                data=df_filtrado, x=var, hue=self.config.variavel_alvo,
                order=contagens.index, palette="Set3",
                edgecolor=".2", linewidth=.5
            )
            
            # Adiciona porcentagens
            total_por_categoria = df_filtrado[var].value_counts()
            for i, p in enumerate(ax.patches):
                altura = p.get_height()
                categoria = p.get_x() + p.get_width()/2
                if i < len(total_por_categoria):  # Evitar anotar as barras de hue
                    total = total_por_categoria.iloc[i % len(total_por_categoria)]
                    porcentagem = f"{(altura/total*100):.1f}%"
                    ax.annotate(porcentagem, (categoria, altura+5),
                                ha='center', va='bottom', fontsize=8)

            plt.title(f"Distribui√ß√£o de {var} por Risco de Cr√©dito", pad=20, fontsize=12)
            plt.xticks(rotation=45, ha='right', fontsize=9)
            plt.xlabel("")
            plt.ylabel("N√∫mero de Clientes", fontsize=10)
            plt.legend(title="Risco", labels=["Bom", "Ruim"])
            plt.tight_layout()
            plt.show()
            
            print(f"""
üìå **AN√ÅLISE DA VARI√ÅVEL CATEG√ìRICA: {var.upper()}**

1. **O QUE O GR√ÅFICO MOSTRA**:
   - Cada barra representa uma categoria de {var}
   - As cores mostram a propor√ß√£o de bons (verde) e maus (laranja) pagadores
   - As porcentagens indicam o percentual de maus pagadores em cada categoria
   - As barras est√£o ordenadas da categoria mais frequente para a menos frequente

2. **COMO INTERPRETAR**:
   - Categorias com maior propor√ß√£o de maus pagadores podem indicar maior risco
   - Categorias dominantes (barras maiores) s√£o as mais comuns na base
   - Diferen√ßas significativas entre categorias sugerem padr√µes importantes

üí° **O QUE OBSERVAR PARA {var}**:
- Quais categorias t√™m maior propor√ß√£o de maus pagadores?
- H√° categorias com risco significativamente maior que outras?
- A distribui√ß√£o entre categorias √© equilibrada ou concentrada?
- Alguma categoria pequena tem risco desproporcionalmente alto?
""")

    def _analisar_correlacoes(self) -> None:
        """Analisa correla√ß√µes entre vari√°veis com heatmap."""
        print("\nüîó AN√ÅLISE DE CORRELA√á√ïES ENTRE VARI√ÅVEIS")
        print("="*50)
        
        # Prepara√ß√£o dos dados
        cols = self.config.variaveis_numericas + [self.config.variavel_alvo]
        matriz_corr = self.df[cols].corr()
        
        # Configura√ß√£o do gr√°fico
        plt.figure(figsize=(14, 12))
        mask = np.triu(np.ones_like(matriz_corr, dtype=bool))
        
        # Heatmap
        sns.heatmap(
            matriz_corr, mask=mask, annot=True, fmt=".2f",
            cmap="coolwarm", center=0, linewidths=.5,
            square=True, cbar_kws={"shrink": .8},
            annot_kws={"size": 9}
        )
        
        plt.title("Matriz de Correla√ß√£o entre Vari√°veis Num√©ricas", pad=20, fontsize=14)
        plt.xticks(rotation=45, ha='right', fontsize=9)
        plt.yticks(fontsize=9)
        plt.tight_layout()
        plt.show()
        
        # Top correla√ß√µes com alvo
        correlacoes_alvo = matriz_corr[self.config.variavel_alvo].drop(self.config.variavel_alvo)
        top_correlacoes = correlacoes_alvo.abs().sort_values(ascending=False).head(5)
        
        print("""
üìå **EXPLICA√á√ÉO DA MATRIZ DE CORRELA√á√ÉO**:

1. **O QUE √â UMA CORRELA√á√ÉO**:
   - Mede como duas vari√°veis variam juntas (de -1 a 1)
   * +1: Aumentam juntas perfeitamente
   * -1: Uma aumenta quando a outra diminui
   * 0: Nenhuma rela√ß√£o linear

2. **COMO LER O HEATMAP**:
   - Cada quadrado mostra a correla√ß√£o entre duas vari√°veis
   - Cores quentes (vermelho): Correla√ß√£o positiva
   - Cores frias (azul): Correla√ß√£o negativa
   - Intensidade da cor = for√ßa da correla√ß√£o
   - Valores num√©ricos mostram a correla√ß√£o exata

3. **FOCO NO RISCO DE CR√âDITO**:
   - A √∫ltima linha/coluna mostra correla√ß√µes com a vari√°vel alvo
   - Correla√ß√µes fortes (>0.3 ou <-0.3) s√£o as mais interessantes
""")
        
        print("\nüèÜ TOP 5 VARI√ÅVEIS MAIS CORRELACIONADAS COM RISCO DE CR√âDITO:")
        print(top_correlacoes.to_string())
        
        print("""
üí° **O QUE OBSERVAR**:
- Quais vari√°veis t√™m maior correla√ß√£o (positiva ou negativa) com o risco?
- H√° vari√°veis altamente correlacionadas entre si (poss√≠vel redund√¢ncia)?
- As correla√ß√µes fazem sentido no contexto de cr√©dito?
- Alguma correla√ß√£o forte √© contra-intuitiva ou inesperada?
""")

    def _detectar_outliers(self) -> None:
        """Identifica outliers usando m√©todo IQR."""
        print("\n‚ö†Ô∏è DETEC√á√ÉO DE VALORES EXTREMOS (OUTLIERS)")
        print("="*50)
        
        resultados = {}
        
        for var in self.config.variaveis_numericas:
            q1 = self.df[var].quantile(0.25)
            q3 = self.df[var].quantile(0.75)
            iqr = q3 - q1
            limite_inf = q1 - 1.5 * iqr
            limite_sup = q3 + 1.5 * iqr
            
            outliers = self.df[(self.df[var] < limite_inf) | (self.df[var] > limite_sup)]
            pct_outliers = (len(outliers) / len(self.df)) * 100
            
            resultados[var] = {
                "Outliers": len(outliers),
                "% do Total": f"{pct_outliers:.1f}%",
                "M√≠nimo Extremo": outliers[var].min() if not outliers.empty else "-",
                "M√°ximo Extremo": outliers[var].max() if not outliers.empty else "-",
                "Limite Inferior": f"{limite_inf:.2f}",
                "Limite Superior": f"{limite_sup:.2f}"
            }
        
        df_outliers = pd.DataFrame(resultados).T.sort_values("Outliers", ascending=False)
        
        print("üìã RESUMO DE OUTLIERS POR VARI√ÅVEL:")
        print(df_outliers.to_string())
        
        print("""
üìå **EXPLICA√á√ÉO SOBRE OUTLIERS**:

1. **O QUE S√ÉO OUTLIERS**:
   - Valores anormalmente altos ou baixos em rela√ß√£o √† maioria
   - Detectados aqui pelo m√©todo IQR (Intervalo Interquartil):
     * Calcula-se o intervalo entre o 1¬∫ (Q1) e 3¬∫ quartil (Q3)
     * Qualquer valor abaixo de Q1-1.5*IQR ou acima de Q3+1.5*IQR √© outlier

2. **IMPLICA√á√ïES**:
   - Podem representar erros de medi√ß√£o ou casos genuinamente extremos
   - Em cr√©dito, altos valores podem indicar maior risco
   - Muitos outliers podem distorcer an√°lises estat√≠sticas

üí° **O QUE OBSERVAR**:
- Quais vari√°veis t√™m muitos outliers?
- Os valores extremos fazem sentido no contexto?
- H√° necessidade de tratamento (remo√ß√£o, ajuste, etc.)?
- Outliers concentrados em bons ou maus pagadores?
""")


# Execu√ß√£o da an√°lise
if __name__ == "__main__":
    print("üîé Iniciando an√°lise de dados de cr√©dito alem√£o...\n")
    analisador = AnalisadorDadosCredito()
    analisador.executar_analise_completa()
```
